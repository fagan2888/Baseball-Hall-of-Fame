```python
x = np.arange(-1.0, 1.0, 0.01)
y0 = 1 + x * 0
y1 = x
y2 = 2 * x * y1 - y0
y3 = 2 * x * y2 - y1
y4 = 2 * x * y3 - y2
y5 = 2 * x * y4 - y3
```

```python
fig, ax = plt.subplots()
ax.plot(x, y0)
ax.plot(x, y1)
ax.plot(x, y2)
ax.plot(x, y3)
ax.plot(x, y4)
ax.plot(x, y5)

ax.set(xlabel='x', ylabel='y',
       title='Chebyshev Polynomials')
ax.grid()

plt.show()
```

```python
x = np.arange(-1.0, 1.0, 0.01)
y0_n = 0 * x
y1_n = x
y2_n = 1 * x * y1_n
y3_n = 1 * x * y2_n
y4_n = 1 * x * y3_n
y5_n = 1 * x * y4_n
```

```python
fig, ax = plt.subplots()
ax.plot(x, y0_n)
ax.plot(x, y1_n)
ax.plot(x, y2_n)
ax.plot(x, y3_n)
ax.plot(x, y4_n)
ax.plot(x, y5_n)

ax.set(xlabel='x', ylabel='y',
       title='Monic Polynomials')
ax.grid()

plt.show()
```

```python
x = np.arange(-0.5, 0.5, 0.01)
y = -(2*x) ** 2+1

```

```python
from matplotlib.patches import Polygon

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set(xlabel='x', ylabel='y',
       title='Center Weighting Win Probability Effects')
ax.grid()

a = -0.05
b = 0.05
# Make the shaded region
ix = np.linspace(a, b)
iy = -(2*ix) ** 2+1
verts = [(a, 0), *zip(ix, iy), (b, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
ax.add_patch(poly)

a = 0.35
b = 0.45
# Make the shaded region
ix = np.linspace(a, b)
iy = -(2*ix) ** 2+1
verts = [(a, 0), *zip(ix, iy), (b, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
ax.add_patch(poly)


plt.show()
```
